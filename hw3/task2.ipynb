```python
import numpy as np
from numpy import exp
def func(x):
    return exp(x[0] + 3 * x[1]) + exp(x[0] - 3 * x[1]) + exp(-x[0])
def f_grad(x):
    return np.array([exp(x[0] + 3 * x[1]), 3 * exp(x[0] + 3 * x[1])]) + np.array([exp(x[0] - 3 * x[1]), -3 * exp(x[0] - 3 * x[1])]) - np.array([exp(-x[0]), 0])

M = 0
for i in range(501):
    for j in range(501):
        x = i / 250 - 1
        y = j / 250 - 1
        if x == y:
            continue
        M = max(M, np.linalg.norm(f_grad(np.array([x, y]))) / np.linalg.norm(np.array([x, y])))
        if M > 1000: 
            print(x, y, M)
M *= 3
print(M)
```

```python
alpha = 1 / M
traj_opt_step = []
x_start = np.array([2, 2.8])
traj_opt_step.append(x_start.copy())
cur_x = x_start.copy()
for i in range(100):
    t = cur_x
    cur_x = cur_x - alpha * f_grad(cur_x)
    if func(t) < func(cur_x):
        alpha /= 2
        cur_x = t
    else:
        traj_opt_step.append(cur_x.copy())

print(traj_opt_step)
```

```python
#Nesterov
alpha = 1 / M
traj_nesterov = []
x_start = np.array([-0.5, -0.5])
traj_nesterov.append(x_start.copy())
cur_x = x_start.copy()
cur_y = x_start.copy()

for i in range(30):
    prev_x = cur_x
    prev_alpha = alpha
    cur_x = cur_y - (f_grad(cur_y) / M)
    print(cur_x)
    alpha = (- prev_alpha ** 2 + np.sqrt(prev_alpha ** 4 + 4 * prev_alpha ** 2)) / 2
    beta = prev_alpha * (1 - prev_alpha) / (prev_alpha ** 2 + alpha)
    cur_y = cur_x + beta * (cur_x - prev_x)
    traj_nesterov.append(cur_x.copy())
print(traj_nesterov)
```
